```python
df_years = pd.DataFrame()
filenames = glob.glob(PATH_TO_DATA + '/*.csv')
for filename in filenames:
    df = pd.read_csv(filename)
    year = filename.split('/')[-1].split('.')[0]
    df_year = pd.DataFrame()
    teams = df['HomeTeam'].unique()
    for team in teams:
        df_team_in = df[(df['HomeTeam']==team)|(df['AwayTeam']==team)]
        df_team_in_home = df[(df['HomeTeam']==team)]
        df_team_in_away = df[(df['AwayTeam']==team)]
        df_team_out = pd.DataFrame()
        df_team_out['Club'] = [team]
        df_team_out['Home_Wins_season'] = len(df_team_in_home[df_team_in_home['FTR']=='H'])
        df_team_out['Away_Wins_season'] = len(df_team_in_away[df_team_in_away['FTR']=='A'])
        df_team_out['Wins_season'] = df_team_out['Home_Wins_season'] + df_team_out['Away_Wins_season']
        df_team_out['Home_Lose_season'] = len(df_team_in_home[df_team_in['FTR']=='A'])
        df_team_out['Away_Lose_season'] = len(df_team_in_away[df_team_in['FTR']=='H'])
        df_team_out['Lose_season'] = df_team_out['Home_Lose_season'] + df_team_out['Away_Lose_season']
        df_team_out['Home_Draw_season'] = len(df_team_in_home[(df_team_in_home['FTR']=='D')])
        df_team_out['Away_Draw_season'] = len(df_team_in_away[(df_team_in_away['FTR']=='D')])
        df_team_out['Draw_season'] = len(df_team_in[df_team_in['FTR']=='D'])
        df_team_out['Points'] = df_team_out['Wins_season']*3 + df_team_out['Draw_season']
        df_team_out['Home_goals_avg'] = df_team_in_home['FTHG'].mean()
        df_team_out['Home_goals_med'] = df_team_in_home['FTHG'].median()
        df_team_out['Home_goals_max'] = df_team_in_home['FTHG'].max()
        df_team_out['Home_goals_mode'] = df_team_in_home['FTHG'].mode()
        df_team_out['Away_goals_avg'] = df_team_in_away['FTAG'].mean()
        df_team_out['Away_goals_med'] = df_team_in_away['FTAG'].median()
        df_team_out['Away_goals_max'] = df_team_in_away['FTAG'].max()
        df_team_out['Away_goals_mode'] = df_team_in_away['FTAG'].mode()    

        df_year = pd.concat([df_year, df_team_out],axis=0,ignore_index=True)
    df_year['Season'] = year
    df_year['Season_start'] = year.split('-')[0]
    df_year = df_year.sort_values(by='Points',ascending=False).reset_index(drop=True).reset_index().rename(columns={'index':'Position'})
    df_year['Position'] += 1
    df_years = pd.concat([df_years, df_year],axis=0,ignore_index=True)
```

```python
filenames_datos = [  PATH_TO_DATA + '/14-15.csv',
                     PATH_TO_DATA + '/15-16.csv',
                     PATH_TO_DATA + '/16-17.csv',
                     PATH_TO_DATA + '/17-18.csv',
                     PATH_TO_DATA + '/18-19.csv',
                     PATH_TO_DATA + '/19-20.csv']
```

```python
def correctDate(s):
    y = s.split('/')[-1]
    if len(y)>2:
        return s
    else:
        return s[:-2]+'20'+s[-2:]
```

```python
#La jornada que surt d'aquí es per calcular quants punts porte, no és la jornada real.
#El merge s'ha de fer per la dependència temporal "Date"
df_jornada = pd.DataFrame()
df_full = pd.DataFrame()
for filename in filenames_datos:
    df = pd.read_csv(filename)
    year = filename.split('/')[-1].split('.')[0]
    df_jornada = pd.DataFrame()
    teams = df['HomeTeam'].unique()
    for team in teams:
        df_equip = df[(df['HomeTeam']==team)|(df['AwayTeam']==team)].reset_index(drop=True).reset_index().rename(columns={'index':'Jornada'})
        df_equip['Season'] = year
        df_equip['Team'] = team
        df_equip['Jornada'] += 1
        df_equip['Result'] = 'D'
        df_equip['GolsFavor'] = 0
        df_equip['GolsContra'] = 0
        df_equip_home = df_equip[(df_equip['HomeTeam']==team)]
        df_equip_away = df_equip[(df_equip['AwayTeam']==team)]
        df_equip_home.loc[df_equip_home['FTHG']>df_equip_home['FTAG'],'Result'] = 'W'
        df_equip_home.loc[df_equip_home['FTHG']<df_equip_home['FTAG'],'Result'] = 'L'
        df_equip_away.loc[df_equip_away['FTHG']<df_equip_away['FTAG'],'Result'] = 'W'
        df_equip_away.loc[df_equip_away['FTHG']>df_equip_away['FTAG'],'Result'] = 'L'
        df_equip_home['GolsFavor'] = df_equip_home['FTHG']
        df_equip_home['GolsContra'] = df_equip_home['FTAG']
        df_equip_away['GolsFavor'] = df_equip_away['FTAG']
        df_equip_away['GolsContra'] = df_equip_away['FTHG'] 
        df_equip_home['Local'] = True
        df_equip_away['Local'] = False
        df_equip = pd.concat([df_equip_home,df_equip_away])
        df_equip = df_equip[['Date','Season','Jornada','Local','Team','GolsFavor','GolsContra','Result']].sort_values(by='Jornada')
        df_equip['Points'] = 0
        df_equip['Season_start'] = year.split('-')[0]
        df_equip.loc[df_equip['Result']=='W','Points'] = 3
        df_equip.loc[df_equip['Result']=='D','Points'] = 1
        df_equip['PercPoints'] = df_equip['Points'].cumsum()/(df_equip['Jornada']*3)
        df_equip['PercPoints'] = df_equip['PercPoints'].shift(1)
        df_full = pd.concat([df_full,df_equip],axis=0,ignore_index=True)   
```

```python
df_final = pd.DataFrame()
for filename in final_filenames:
    df = pd.read_csv(filename)
    df_total = pd.DataFrame()
    dic = {'Home': True, 'Away': False}
    df['Season'] = filename.split('/')[-1].split('.')[-0]
    df = df[['Season','Date','HomeTeam','AwayTeam','B365H','B365D','B365A','FTR']].copy()
    #display(df[df.isna().any(axis=1)])
    df['Date'] = df['Date'].apply(correctDate)
    df_full['Date'] = df_full['Date'].apply(correctDate)
    df['Datetime'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
    df_full['Datetime'] = pd.to_datetime(df_full['Date'], format='%d/%m/%Y')
    df['Season_start'] = df['Season'].apply(lambda x: int(x.split('-')[0])).astype(int)
    df_years['Season_start'] = df_years['Season_start'].astype(int)
    for idx, row in df.iterrows():
        df_fila = pd.DataFrame()
        df_fila['B365H'] = [row['B365H']]
        df_fila['B365D'] = row['B365D']
        df_fila['B365A'] = row['B365A']
        df_fila['FTR'] = row['FTR']
        df_fila['Date'] = row['Date']
        df_fila['HomeTeam'] = row['HomeTeam']
        df_fila['AwayTeam'] = row['AwayTeam']
        for local in ['Home','Away']:
            df_jornada = pd.DataFrame()
            df_season = df_full[(df_full['Team']==row[local+'Team'])&(df_full['Season']==row['Season'])&(df_full['Datetime']<row['Datetime'])].copy()
            valid = (len(df_season)>0)
            if valid:
                #This season
                df_jornada[local+'TeamPercPoints'] = [df_full[(df_full['Date']==row['Date'])&(df_full['Team']==row[local+'Team'])].copy()['PercPoints'].values[0]]
                df_jornada[local+'TeamGolsFavorMean'] = df_season['GolsFavor'].mean()
                df_jornada[local+'TeamGolsFavorMedian'] = df_season['GolsFavor'].median()
                df_jornada[local+'TeamGolsFavorMax'] = df_season['GolsFavor'].max()
                df_jornada[local+'TeamGolsContraMean'] = df_season['GolsContra'].mean()
                df_jornada[local+'TeamGolsContraMedian'] = df_season['GolsContra'].median()
                df_jornada[local+'TeamGolsContraMax'] = df_season['GolsContra'].max()
                #"Last" season
                last_season_start = df_years[(df_years['Club']==row[local+'Team'])&(df_years['Season_start']<row['Season_start'])]['Season_start'].max()
                df_lastseason = df_years[(df_years['Club']==row[local+'Team'])&(df_years['Season_start']==last_season_start)].copy()
                df_lastseason = df_lastseason.drop(['Club', 'Season','Season_start'], axis=1)
                for col in df_lastseason.columns.values:
                    df_lastseason = df_lastseason.rename(columns={col:(local+'Team_'+col)})
                df_jornada = pd.concat([df_jornada.reset_index(drop=True), df_lastseason.reset_index(drop=True)], axis= 1)
                #Fifa concat
                df_fifa_fila = df_fifa_club[(df_fifa_club['Club']==row[local+'Team'])&(df_fifa_club['Season']==row['Season'])].copy()
                df_fifa_fila = df_fifa_fila.drop(['Club', 'Season'], axis=1)
                for col in df_fifa_fila.columns.values:
                    df_fifa_fila = df_fifa_fila.rename(columns={col:(local+'Team_'+col)})
                df_jornada = pd.concat([df_jornada.reset_index(drop=True), df_fifa_fila.reset_index(drop=True)], axis= 1)

                df_fila =  pd.concat([df_fila.reset_index(drop=True), df_jornada.reset_index(drop=True)], axis= 1)

                if (len(df_lastseason)!=1):
                    print('OJO, lequip no té una season anterior!!')
        if (valid):
            df_total = pd.concat([df_total,df_fila],axis=0,ignore_index=True)  
        
    #df_total = df_total.drop(columns=['Date','HomeTeam'])
    df_final =  pd.concat([df_final.reset_index(drop=True), df_total.reset_index(drop=True)], axis= 0)
    print(filename)
    print(len(df_total))
    print(len(df_total.dropna()))
```
